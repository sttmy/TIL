```python
# https://www.kaggle.com/anthonypino/price-analysis-and-linear-regression

import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/data_houseprice.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
X= df.loc[ : , ['Rooms', 'Distance', 'Bedroom2','Bathroom', 'Car', 'Landsize', 'BuildingArea','Propertycount']]
y = df["Price"]
df2=pd.concat([X,y],axis=1)

# 결측값처리
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="mean") # 평균
df3 = pd.DataFrame(imputer.fit_transform(df2),columns=df2.columns)
df3.head(3)
```

```python
df3.isnull().sum()
```

```python
cols=df3.columns[:-1]
X=df3[cols]
y=df3['Price']
```

```python
#다중 회귀분석
import statsmodels.api as sm
model = sm.OLS(y, X)     # sm.OLS(y~x) 같은 것임
result = model.fit()
print("회귀계수\n",result.params)
print("R-squared\n",result.rsquared)
print("P values\n",result.pvalues)
#요약 결과 출력
result.summary()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#데이터셋을 나눔(학습용:검증용 = 8:2)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
model = LinearRegression().fit(X_train, y_train)
print("학습용:",model.score(X_train, y_train))
print("검증용:",model.score(X_test, y_test))
```
