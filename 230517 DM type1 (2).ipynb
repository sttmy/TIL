14.각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
t = channel[(channel.ct >= '2021-10-03 03:00:00') & (channel.ct <= '2021-11-01 15:00:00')]
```

```python
t
```

```python
inc = []
for i in range(0, len(t.channelname.unique())):
    oc = t[t.channelname == t.channelname.unique()[i]].iloc[0].subcnt
    nov = t[t.channelname == t.channelname.unique()[i]].iloc[-1].subcnt
    inc.append(nov - oc)
```

```python
pd.DataFrame(list(zip(t.channelname.unique(), inc)), columns = ['channelname','subcnt증가량'])
```

15.각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다. 공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라

```python
video.head(2)
```

```python
video.info()
```

```python
gong1 = video[video['videoname'] == video['videoname'][0]]
gong1
```

```python
(gong1.ct[1] - gong1.ct[0] <= dt.timedelta(minutes = 5) ) | (gong1.ct[1] - gong1.ct[0] >= dt.timedelta(minutes = 20) )
```

```python
ct_diff = [False]
for i in range(0, len(gong1)):
    try:
        diff = gong1.ct[i+1] - gong1.ct[i]
        ct_dif = (diff <= dt.timedelta(minutes = 5) ) | (diff >= dt.timedelta(minutes = 20)) 
        ct_diff.append(ct_dif)
    except:
        pass
ct_diff
```

```python
len(ct_diff), len(gong1)
```

```python
gong1[ct_diff]
```

```python
gong1.iloc[[9,10,417,418,720,721,722,1635,1636]]
```

```python
gong1_ecpt = gong1.sort_values('ct').reset_index(drop = True)
gong1_ecpt 
```

```python
gong1_ecpt[ ( gong1_ecpt.ct.diff(1) >= dt.timedelta(minutes = 20) ) | 
          ( gong1_ecpt.ct.diff(1) <= dt.timedelta(minutes = 5) ) ]
```
