#### 객체지향 언어 ex) C++, python, JAVA

- "누가 어떤 일을 할 것인가?"

- ①추상화: 객체들의 공통적인 특징을 도출
- ②캡슐화: 데이터와 기능을 하나로 묶어 관리, 코드가 묶여 있고 오류 없어 편리
- ③상속성: 하나의 클래스가 가진 특징(함수,데이터)을 다른 클래스가 그대로 물려받음
- ④다형성: 동일한 명령의 해석을 연결된 객체에 의존 
<br> 오버라이딩: 부모클래스의 매소드와 같은 이름을 사용하며, 매개변수도 같도록 내부 소스를 재정의
<br> 오버로딩: 같은 이름의 함수를 여러개 정의한 후, 매개변수를 다르게 하여 같은 이름을 호출하여 사용
- ⑤동적 바인딩

#### 객체 지향 프로그래밍의 장점

- 소프트웨어의 생산성 향상: SW작성 부담이 줄여짐, 코드를 재사용하기 쉬움, 업그레이드 및 디버깅 쉬움
- 실세계에 대한 쉬운 모델링
- 보안성 향상

#### 객체 지향 프로그래밍의 단점

- 느린 실행 속도: 필요한 메모리양 증가

## ★ 프로토콜 Protocol: 컴퓨터끼리 정보를 주고 받는 규칙이나 약속

1. 프로토콜
    - 인터넷: TCP(Transmission Control Protocol)와 IP(Internet Protocol)
        - TCP: 한 기기에서 다른 기기로 데이터를 전송
        - IP: 데이터의 조각을 최대한 빨리 대상 IP주소로 전송
        
    - 웹: HTTP(HyperText Transfer Protocol)

    - 파일을 주고 받을 때: FTP(File Transfer Protocol)
    
    - 메일을 주고받을 때: SMTP(Simple Mail Transfer Prococol)



2. 계층
    - TCP/IP 4계층: 프로토콜 적용 특정 조건
    - Transport Layer(4 Layer): 송수신자의 논리적 연결을 담당. 연결과 데이터를 얼마나 보냈는지 얼마나 받았는지 제대로 받았는지 확인
    - Network Layer(3 Layer): IP가 활용되는 부분, 경로와 목적지를 찾아줌. Routing. 대역이 다른 IP목적지를 찾아갈 수 있도록 돕는 역할
    
    
    - 응용계층 Application Layer: TCP/IP 프로토콜 범위에 포함되어 있음
    
    - 전송계층 Transport Layer: TCP(데이터전송, 연결형서비스), UDP(비연결형서비스)
    
    - 인터넷계층 Internet Layer: 네트워크 간 데이터 패킷의 전송을 관리. ARP, IGMP, ICMP,...
    
    - 데이터링크계층 Datalink Layer: 데이터가 원하는 IP주소에 도달 또는 기기에 연결되어있는지 확인
