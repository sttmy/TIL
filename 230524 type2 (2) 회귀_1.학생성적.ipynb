## Feature Engineering, 표준화(스케일링)

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
x_train[['age','G1','G2']] = Scaler.fit_transform(x_train[['age','G1','G2']])
x_test[['age','G1','G2']] = Scaler.fit_transform(x_test[['age','G1','G2']])
```

## 원-핫 인코딩

```python
# 원핫인코딩: long-type으로 구성된 변수를 wide-type으로 목록화하여 변수의 차원을 늘림
# 개별로 해당 컬럼이 맞으면 1, 아니면 0의 이진수를 가짐
# pandas.get_dummis를 사용
# 문제점: multicollinearity 독립변수간의 강한 상관관계
```

```python
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)
```

```python
from sklearn.model_selection import train_test_split
x_trn, x_ts, y_trn, y_ts = train_test_split(x_train, y_train, 
                                          test_size=0.3, random_state=100)
```

## 모델링

```python
# 선형회귀분석
from sklearn.linear_model import LinearRegression
modelLR = LinearRegression()
```

```python
modelLR.fit(x_trn, y_trn)
modelLR.score(x_trn, y_trn)
```

```python
# 랜덤포레스트
from sklearn.ensemble import RandomForestRegressor
modelRF = RandomForestRegressor()
modelRF.fit(x_trn, y_trn)
modelRF.score(x_trn, y_trn)
```

## 모델 학습 및 예측, 평가

```python
y_pred_valLR = modelLR.predict(x_ts)
y_pred_valRF = modelRF.predict(x_ts)

from sklearn.metrics import r2_score
print("LR r2_score", r2_score(y_ts, y_pred_valLR))
print("RF r2_score", r2_score(y_ts, y_pred_valRF))
```

# https://www.kaggle.com/code/minjunim/py-rf-xgb-gb-r2-score-0-816

```python
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv")
y_test = pd.read_csv('y_test.csv')
```

```python
y_train = y_train['G3']
y_test = y_test['G3']
```

```python
# Scaling

ncols = x_train.select_dtypes(exclude="object").columns
ocols = x_train.select_dtypes(include="object").columns
```

```python
print(ocols), print(ncols)
```

```python
# Standard Scaling
from sklearn.preprocessing import StandardScaler

print ('before scaling \n', x_train.head())
print ('------------------------------')

scaler = StandardScaler()
x_train[ncols] = scaler.fit_transform(x_train[ncols])
x_test[ncols] = scaler.fit_transform(x_test[ncols])

print ('after scaling \n', x_train.head())
```

```python
# Encoding

from sklearn.preprocessing import LabelEncoder

print ('before encoding \n', x_train.head())
print ('------------------------------')

for col in ocols:
    le = LabelEncoder()
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.fit_transform(x_test[col])
    
print ('after encoding \n', x_train.head())
```

```python
### Splitting Data

# 컬럼 순서 맞추기
cols = x_train.columns
x_test = x_test[cols]
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state = 42)
x_tr.shape, x_val.shape, y_tr.shape, y_val.shape
```

```python
# Modeling
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor, XGBRFRegressor

from sklearn.metrics import mean_squared_error, r2_score
```

```python
rf = RandomForestRegressor(random_state = 156)
gb = GradientBoostingRegressor(random_state = 156)
xgb = XGBRegressor(random_state = 156)
models = [rf, gb, xgb]
```

```python
models
```

```python
# 최적의 파라미터 값 찾아보기

from sklearn.model_selection import GridSearchCV

params = {'n_estimators':[100,200,300,400], 'max_depth':[1,2,3,4,5]}

# GridSearch
for model in models :
    gs = GridSearchCV(model, params, cv=3, scoring ="r2", n_jobs=4)
    gs.fit(x_tr, y_tr)
    
    print(f"="*10)
    print(f"model : {model}")
    print(f"params : {gs.best_params_}")
    print(f"score : {gs.best_score_}")
```

```python
# model : RandomForest

model_rf = RandomForestRegressor(random_state=156, max_depth=3, n_estimators=200)
model_rf.fit(x_tr, y_tr) 

pred = model_rf.predict(x_test)
```

```python
# model score : train
pred_tr = model_rf.predict(x_tr)

# model score : Validation
pred_val = model_rf.predict(x_val)

print('rmse_train :', np.sqrt(mean_squared_error(y_tr, pred_tr)))
print('r2_score_train :', r2_score(y_tr, pred_tr))

print("="*19)

print('rmse_val :', np.sqrt(mean_squared_error(y_val, pred_val)))
print('r2_score_val :', r2_score(y_val, pred_val))
```

```python
# model score : test

print('rmse :', np.sqrt(mean_squared_error(y_test, pred)))
print('r2_score :', r2_score(y_test, pred))
```

```python
# saving and submission
submit = pd.DataFrame({"StudentID": x_test_id, "G3" : pred})

print(submit.head())

submit.to_csv("result.csv", index = False)

# check = pd.read_csv("result.csv")
# check.head()
```
