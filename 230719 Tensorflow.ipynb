## I. 딥러닝
### A. 배경

### 1) 머신러닝/인공지능
- 인간이 코딩 등으로 사전에 정의 내린 행동이 아닌 행동을 하도록 하는 것에 가까운 개념(Arthur Samuel, 1959). Field of study that gives computers the ability to learn without being explicitly programmed (no if –then )
- 스팸필터, 상품추천, 자율주행차 등의 다양한 분야에서 활용되고 있음
- 머신러닝은 러닝, 즉 “학습”의 개념이 포함됨

### 2) 학습
- 머신러닝을 코딩한다는 것은 결국 “학습”하는 방법을 코딩한다는 것
- 어떻게 학습을 시킬 것인가 = 어떻게 시행착오를 거치게 할 것인가?
- ① 어떤 목표를 정해 놓고 목표를 달성하게 할 경우  의도된 결과를 도출하도록하는 것
- ② 손실 함수, 비용 함수(cost function, loss function): 학습을 통해 그 값을 최소로 만드는 것을 목표로 하는 함수

### 3) Supervised vs. Unsupervised
- ① 지도학습(Supervised learning)
    - 학습 데이터에 답(labeled data)이 함께 제공
    - 고양이 그림에 고양이라고 표시해 주는 것
    - 대부분의 머신러닝 알고리즘이 해당됨
- ② 비지도학습(Unsupervised learning)
    - 학습데이터에 답이 없음(unlabeled data)
    - 미리 답을 줄 수 없는 데이터
    - 일부 머신러닝기법에서 사용

### 4) 예측과 분류(Prediction/Regression, Classification)
 예측(Prediction, Regression)

답을 알고 있는 데이터로 학습을 시킨 후 새로운 데이터가 들어왔을 때 가장유사한 결과값을 예상하도록 하는 것

2) 분류(Classification)

레이블이 있는 데이터로 분류하는 것을 학습시킨후 레이블이 없는 새로운 데이터가 들어왔을 때 레이블을 부여하는 것

### C. Tensorflow(http://www.tensorflow.org)
- 구글에서 만든 오픈소스 머신러닝 프레임웍
- 다양한 언어를 제공하며 파이썬이 가장 많이 사용됨

### 1) 특징
- ① 기본적으로 텐서를 활용한 그래프 수치 연산을 하는 도구
- ② 수학적인 의미에서의 그래프 : 노드와 엣지로 구성된 기하 모형
- ③ 노드(node) : 연산 및 데이터를 정의하는 것
- ④ 엣지(edge) : 노드들을 연결하는 것(데이터의 흐름)
- ⑤ 텐서(Tensor) : 다차원 데이터 배열
- ⑥ 텐서플로 – 텐서가 노드에서 연산되고 엣지를 통해 돌아다닌다

### 2) 기본개념
- ① 오퍼레이션(Operation) 그래프 상의 노드 하나 이상의 텐서를 받을 수 있음 계산을 수행하고 결과를 하나 이상의 텐서로 리턴
- ② 텐서(Tensor) 일종의 다차원 배열(행렬) 내부적으로 모든 데이터는 텐서를 통해 표현됨 그래프 내의 오퍼레이션 간에는 텐서만이 전달됨
- ③ 변수(Variables) 그래프의 실행시 파라미터(가중치,편향)를 저장하고 갱신하는데 사용됨 메모리 상에서 텐서를 저장하는 버퍼 역할
    - variable : 학습시키면서 최적화하는 변수의 용도로 사용됨. 학습과정에서 지속적으로 바뀔 수 있는 중요한 값
    - constant : 상수
    - 변수의 선언: tf.Variable(초기값, 타입)
    - 변수의 초기화(텐서플로 2.0에서 없어짐): tf.global_variables_initializer()를 통해 초기화할 수 있음. 
    변수는 반드시 초기화를 해야함. 초기화는 노드에 값을 입력한다는 의미
- ④ 텐서플로 자료형의 종류
    - tf.float16 : 표준16비트 부동소수점
    - tf.float32 : 표준32비트 부동소수점
    - tf.float64 : 표준64비트 부동소수점
    - tf.bfloat16 : 텐서플로만의 16비트 truncated 부동소수점
    - tf.complex64 : 표준64비트 복소수
    - tf.complex128 : 표준64비트 복소수
    - tf.int8 : 표준8비트 정수
    - tf.uint8 : 표준8비트 무부호정수(양수정수)
    - tf.uint16 : 표준16비트 무부호정수(양수정수)
    - tf.int16 : 표준16비트 정수
    - tf.int32 : 표준32비트 정수
    - tf.int64 : 표준64비트 정수
    - tf.bool : 불린(Boolean)
    - tf.string : 문자열
    - tf.qint8 : 텐서플로만의 Quantized 8비트정수
    - tf.quint8 : 텐서플로만의 Quantized 8비트무부호 정수(양수)

```python
# 텐서플로 변수선언
W = tf.Variable([0.3], dtype = tf.float32)    # 가중치
b = tf.Variable([-0.3], dtype = tf.float32)   # 편향

def linear_model(x):
    return tf.multiply(W, tf.add(x,b))      # W * (x+b)

x = [1., 2., 3., 4.]

# x값에 따라 linear_model계산
print(linear_model(x).numpy())    # 0.3*0.7  0.3*1.7  0.3*2.7  0.3*3.7
print(linear_model(x))
```

### 2. 케라스(Keras)
- 파이썬으로 개발된 오픈 소스 신경망 라이브러리
- Deeplearning4j, 텐서플로 등의 기존 딥러닝 라이브러리를 기반으로 더 직관적이고 간결한 코드로 빠른 실험이 가능하도록 구현됨
- 텐서플로 2로 버전업이 되면서 케라스의 중요성이 더욱 강조됨
- 공식 사이트(http://keras.io)

### 1) 개요
- 파이썬으로 구현된 쉽고 간결한 딥러닝(deep learning) 라이브러리
- 구글의 엔지니어인 프랑소와 숄레(François Chollet)가 2015년 3월에 발표
- 최근 버전 : 2022년 6월 현재 2.6.0
- 내부적으로는 텐서플로우(tensorflow) 엔진이 구동되지만 직관적인 API로 쉽게 딥러닝 실험을 할 수 있도록 지원함

### 2) 케라스의 주요 특징
- ① 모듈화(Modularity) : 독립적인 모듈들을 조합하여 구현
- ② 최소주의(Minialism) : 각 모듈은 짧고 간결
- ③ 쉬운 확장성 : 클래스나 함수로 모듈을 쉽게 추가할 수 있음
- ④ 파이썬 기반 : 별도의 설정이 필요없음

### 3) 케라스를 이용한 실습 순서
- ① 데이터셋 생성
    - 데이터로부터 학습용, 검증용 데이터셋 구분
    - 딥러닝 모델의 학습 및 검증을 할 수 있도록 케라스에서 요구하는 형태로 변환을 해야 함
- ② 모델 구성
    - Sequence 모델을 생성한 후 필요한 Layer를 추가하여 구성
- ③ 모델 학습과정 설정
    - 손실 함수 및 최적화 방법 정의
    - 모델 컴파일
- ④ 모델 학습
    - 학습용 데이터를 이용하여 모델 학습
- ⑤ 학습과정 출력
    - 학습 시 학습용, 검증용 데이터셋의 손실 및 정확도 측정
- ⑥ 모델 평가
    - 검증용 데이터셋으로 학습한 모델을 평가
    - 쓸만하면 실제 적용, 아니면 재학습
- ⑦ 새로운 데이터 사용하여 값을 예측

```python
import numpy as np
np.random.seed(7)

#데이터 로딩, 시작36행은 건너뜀
data=np.genfromtxt("Data/x09.txt",skip_header=36)
data[:10]    # Blood / fat / content
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xdata = np.array(data[:, 2], dtype = np.float32)
ydata = np.array(data[:, 3], dtype = np.float32)
zdata = np.array(data[:, 4], dtype = np.float32)
fig = plt.figure(figsize = (12,12))
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(xdata, ydata, zdata)
ax.set_xlabel('weight')
ax.set_ylabel('age')
ax.set_zlabel('blood sugar')
ax.view_init(15, 15)   # 출력 각도
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop

# 데이터를 케라스에 맞게 변환
x_data = np.array(data[:,2:4], dtype = np.float32)  # 독립변수
y_data = np.array(data[:,4], dtype = np.float32)  # 종속변수

# 회귀분석 모델 생성
model = Sequential()    # 신경망모형
model.add(Dense(1, input_shape = (2,)))
    # Dense: 밀집. 값이 모두 채워진 행렬    
    # Sparse: 희소. non-zero인 위치와 값만 저장한 행렬이 희소행렬
    # Dense(output, input_shape: 입력데이터 차원(독립변수: 나이와 체중 변수2개), 1차원)
    # hidden 없는 단순 신경망 
    # (input) 나이 * weight1 + 체중 * weight2   (output) 혈당수치
    # 파라미터: weight1, weight2, bias

rmsprop = RMSprop(learning_rate = 0.01)  # 학습률
model.compile(loss = 'mse', optimizer = rmsprop)
    # loss 손실함수 : 인공신경망에서 학습을 통해 최적의 가중치 매개변수를 결정하기 위한 기준으로 사용하는 함수(오차를 계산하는 함수)
    ### 평균제곱오차(Meam Squared Error, MSE), 회귀분석에서 많이 사용하는 손실함수
    # optimizer 최적화함수 : 오차를 줄이기 위해 weight, bias 를 수정하는 함수: adam, rmsprop
    ### https://keras.io/ko/optimizers/  옵티마이저 참조
    # learning_rate 학습률 : 작으면 미분을 자세하게 하게되고 시간이 오래걸림. 간격을 크게하면 건너뛰는값이 많게 됨

# 손실함수의 결과값을 가장 작게 만드는 것이 신경망 학습의 목표
# 손실함수의 결과값을 작게 만들기 위해 가중치 매개변수를 조절해가는 과정이 학습과정

model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes = True)
```

```python
hist = model.fit(x_data, y_data, epochs = 1000)
hist.history.keys()
```

```python
plt.plot(hist.history["loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["train"],loc="upper left")
plt.show()
```

```python
# 새로운 데이터로 예측: 체중 100kg, 40세

model.predict(np.array([100,40]).reshape(1,2))   
    # reshape 행, 열: 1차원 배열을 1행 2열의 2차원 배열로 변환  
model.predict([[100,40]])  # 2차원으로 직접 입력
```

```python
# c:/사용자/본인계정/.keras/datasets 디렉토리에 다운로드됨
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()

#학습용 404개, 검증용 102개, 13개의 변수
print(X_train.shape)
print(X_train.shape[1])
```

```python
y_train[0:10]   # 주택가격
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.DataFrame(X_train).boxplot()
plt.show()
```

```python
# 스케일링 : 변수 간 단위로 인한 편차를 줄이기 위해

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # 평균 0, 표준편차 1로 만드는 스케일러
scaler.fit(X_train)        # 패턴분석
X_train_scaled = scaler.transform(X_train)  # 변환
X_test_scaled = scaler.transform(X_test)    # 변환

pd.DataFrame(X_train_scaled).boxplot()
plt.show()
```

```python
### 활성화 함수 : 입력데이터를 적절하게 처리하여 출력데이터를 만들어내는 함수
# input data -> {activation function} -> output data
# 활성화 함수의 종류 : 탄젠트(thanh) 함수, 시그모이드(sigmoid) 함수, ReLU 함수
# ReLU(Rectified Linear Unit) 함수: 탄젠트 함수나 시그모이드 함수보다 학습시간이 빨라서 최근에는 많이 사용되고 있음

from keras import models, layers
def build_model():
    model = models.Sequential()
    # 64개의 노드를 가진 은닉층 2개의 신경망
    model.add(layers.Dense(64, activation='relu',
                           input_shape=(X_train.shape[1],)))    # input node 13개, hidden-1 node 64개
    model.add(layers.Dense(64, activation='relu'))             # hidden-2 node 64개
    
    # 출력층(1개의 노드, 하나의 값을 예측하는 회귀분석의 경우 활성화 함수가 없는 선형층을 사용함)
    model.add(layers.Dense(1))        # output node 1개, activation 디폴트: 'linear' 항등함수
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
        # mse 손실 함수를 사용하여 컴파일(평균 제곱 오차: 예측값과 실제값의 거리의 제곱)
        # metrics 평가지표, 검증용 데이터셋으로 평가하는 기준
        # mae : 평균 절대 오차, 예측값과 실제값의 거리의 절대값
    return model

model = build_model()
model.summary()
```

```python
from keras.utils import plot_model  
plot_model(model, show_shapes=True)
```

```python
# 학습과정을 그림으로 그려주는 코드

import matplotlib.pyplot as plt
def train(x_train, y_train, epochs=21):
    for epoch in np.arange(1, epochs):
        hist = model.fit(x_train, y_train)         # 학습
        plt.plot(y_train, c='r', label='real')   # 실제값
        plt.plot(model.predict(x_train), c='b', label='pred')    # 예측값(변화하면서 관찰)
        plt.legend()
        mae = hist.history['mae'][0]           # 평균절대오차를 title에 표시 
        plt.title(f"Epoch {epoch:02d}, mae {mae:.2f}")
        plt.savefig(f"Data/img/plot_{epoch:02d}.png")   # 이미지로 저장
        plt.clf()

train(X_train_scaled, y_train)
```

### 예제) IRIS

```python
# 붓꽃 데이터셋 로딩
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris['species'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
X = iris.iloc[:, :4].values   # 독립변수   iloc 인덱스기반 [행, 열]  start:stop:step
y = iris.iloc[:, 4].values   # 종속변수    loc 필드명기반
print(X[:5])
print(y[:5])
```

```python
# 정수인코딩
encoder = LabelEncoder()   # 문자열에 label을 붙임
y1 = encoder.fit_transform(y)
print(y1[:5])
```

```python
# 다중분류이므로 원핫인코딩
y2 = pd.get_dummies(y1).values  # 종속변수 원핫인코딩
print(y2[:5])
```

```python
# 학습용, 검증용 데이터셋 구분
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y2, test_size=0.2, stratify=y,random_state=10)
```

```python
# 모델 생성

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

model = Sequential()
model.add(Dense(64,input_shape=(4,),activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(3,activation='softmax'))   # 다중분류: softmax / 이진분류: sigmoid
sgd = SGD(learning_rate=0.01)  
    # SGD: 확률적 경사하강법
    # 경사하강법은 모든 sample을 일일이 다 검사함
    # 확률적 경사하강법은 전체가 아니라 한 사이클에 random으로 하나만 뽑아서 기울기 계산, 오차 작은 방향으로 움직임
    # 학습률: 미분 간격. 작을수록 자세히 학습시킴(속도가 느림). 클수록 보폭이 커서 최저점 지나칠 수 있음
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
    # loss손실함수 회귀: mse, 분류: (다중)categorical_crossentropy (이진)binary_crossentropy
    # metrics평가기준
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes = True)
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau, CSVLogger

# Earlystopping 조기학습종료 판정
es = EarlyStopping(monitor = 'val_loss', patience= 10)
    # val_loss: validation data의 loss를 기준으로 결정
    # patience = 10 : 10번 참았는데 개선 안되면 끝내라

# ModelCheckpoint: 모델 히스토리 남김
mc = ModelCheckpoint('Data/result/iris_best.h5', save_best_only=True)
    # 모델의 확장자는 .h5를 사용 (파라미터, 가중치 등 저장)
    # save_best_only = True: 제일 좋은 모델만 저장해라. 학습 과정에서 만들어진 최적의 파라미터 조합으로 모형을 저장

# ReduceLROnPlateau 학습이 잘 안될 때 학습률을 동적으로 조절하는 옵션
rlr = ReduceLROnPlateau(factor = 0.1, patience = 5)  # 기존학습률 * factor(배수)
    # factor: 기존학습률(0.1) * 0.1해라 : 학습률이 줄어듦
    # 5번 했는데 개선이 안되면 학습률을 줄여라.

# CSVlogger: 학습과정의 주요 history를 csv로 저장
csvlogger = CSVLogger('Data/result/csv')

hist = model.fit(X_train, y_train, validation_split = 0.2, epochs = 100, callbacks = [es, mc, rlr, csvlogger])
    # validation data : 학습에 직접 참여하지 않지만 파라미터 튜닝하고 조기학습종료여부 결정에 사용
    # train : validation : test = 6:2:2
    # callbacks : 학습 한번 끝나면 한 번 갔다옴 ?

# epoch 미니배치
# train은 수치가 개선될 수밖에 없으므로 val_loss, val_accuracy 등을 관찰
```
