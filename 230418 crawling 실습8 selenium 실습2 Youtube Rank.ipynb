## tch

```python
url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube' 
driver = webdriver.Chrome()
driver.get(url)
time.sleep(2)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
trs = soup.select('tr.aos-init')
len(trs)
```

```python
tr = trs[0]
tr.select_one('.rank').get_text().strip()  # 순위
tr.select_one('.category').get_text().strip().[1:-1]  # 카테고리
tr.select_one('.subject a').get_text().strip() # 채널명
tr.select_one('.subscriber_cnt').get_text().strip() # 구독자수
tr.select_one('.view_cnt').get_text().strip() # 조회수
tr.select_one('.video_cnt').get_text().strip() # 비디오수
```

```python
# 첫페이지 크롤링
lines = []
for tr in trs:
    rank = tr.select_one('.rank').get_text().strip()  # 순위
    category = tr.select_one('.category').get_text().strip()[1:-1]  # 카테고리
    channel = tr.select_one('.subject a').get_text().strip() # 채널명
    subscriber = tr.select_one('.subscriber_cnt').get_text().strip() # 구독자수
    view = tr.select_one('.view_cnt').get_text().strip() # 조회수
    video = tr.select_one('.video_cnt').get_text().strip() # 비디오수
    lines.append([rank, category, channel, subscriber, view, video])
dd = pd.DataFrame(lines, columns = ['순위','카테고리','채널명','구독자수','조회수','비디오수'])
dd.tail()
```

```python
# 숫자로 바꿔주는 함수 만들기
def convert_unit(s):
    s = s.replace('억','').replace('개','').replace(',','').replace('만','0000')
    return f'{int(s):,d}'   # 쉼표 넣어줌
```

```python
convert_unit('46억2878만')
```

```python
for column in dd.columns[3:]:
    dd[column] = dd[column].apply(convert_unit)      # 함수적용
```

```python
driver = webdriver.Chrome()
lines = []
for page in tqdm(range(1,11)):
    url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=' + f'{page}'
    driver.get(url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    trs = soup.select('tr.aos-init')
    
    for tr in trs:
        rank = tr.select_one('.rank').get_text().strip()  # 순위
        category = tr.select_one('.category').get_text().strip()[1:-1]  # 카테고리
        channel = tr.select_one('.subject a').get_text().strip() # 채널명
        subscriber = tr.select_one('.subscriber_cnt').get_text().strip() # 구독자수
        view = tr.select_one('.view_cnt').get_text().strip() # 조회수
        video = tr.select_one('.video_cnt').get_text().strip() # 비디오수
        lines.append([rank, category, channel, subscriber, view, video])

        
dd = pd.DataFrame(lines, columns = ['순위','카테고리','채널명','구독자수','조회수','비디오수'])
for column in dd.columns[3:]:
    dd[column] = dd[column].apply(convert_unit)      # 함수적용
dd
```

```python
dd.to_csv('data1/youtuberank_tch.csv',index = False)
```
