```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False   # 마이너스 표시
plt.rc('font', family = 'Malgun Gothic')
```

```python
pd_data = pd.read_csv('data1/Danawa_preprocessing.csv', encoding = 'UTF-8')
pd_data
```

```python
handy = pd_data[pd_data['카테고리']=='핸디스틱청소기']
handy
```

```python
handy.isnull().sum()
```

```python
# 흡입력 기준 sort
top_list = handy.sort_values(['흡입력'], ascending = False)
top_list
```

```python
# 사용시간 기준 sort
top_list2 = handy.sort_values(['사용시간'], ascending = False)
top_list2
```

```python
# 사용시간 + 흡입력 기준 sort
top_list3 = handy.sort_values(['사용시간', '흡입력'], ascending = False)
top_list3
```

```python
# 평균값 정리

price_mean_value = handy['가격'].mean()
suction_mean_value = handy['흡입력'].mean()
use_time_mean_value = handy['사용시간'].mean()
price_mean_value, suction_mean_value, use_time_mean_value
```

```python
# 가성비 좋은 제품 선택
# : 가격이 평균보다 낮고, 흡입력과 사용시간이 평균보다 높으면 가성비가 좋다고 정의

condition_data = handy[ (handy['가격'] <= price_mean_value) & 
                      (handy['흡입력'] >= suction_mean_value) &
                       (handy['사용시간'] >= use_time_mean_value) ]
    # 조건: handy[가격] <= 가격평균 & handy[흡입력] >= 평균 & handy[사용시간] >= 평균
condition_data
```

```python
condition_data.shape
```

```python
# 데이터 시각화

chart_data = handy.dropna(axis=0)
len(chart_data)
```

```python
# 흡입력, 사용시간의 최대값과 최소값 정리

suction_max_value = chart_data['흡입력'].max()
suction_min_value = chart_data['흡입력'].min()
use_time_max_value = chart_data['사용시간'].max()
use_time_min_value = chart_data['사용시간'].min()

suction_max_value, suction_min_value, use_time_max_value, use_time_min_value
```

### 청소기 성능 시각화

```python
# 청소기 성능 시각화
plt.figure(figsize = (12, 8))
plt.title("핸디스틱 청소기 성능표")
sns.scatterplot(x = '흡입력', y = '사용시간', size = '가격', hue=chart_data['회사명'],
                data= chart_data, sizes = (10,1000), legend = False)
# 중간 평균선그리기
plt.plot([0, suction_max_value], [use_time_mean_value, use_time_mean_value],   
         'r--', lw = 1)   # 'r--': 빨간색, lw: line width
plt.plot([suction_mean_value, suction_mean_value], [0, use_time_max_value], 
         'r--', lw = 1)
plt.show()
```
