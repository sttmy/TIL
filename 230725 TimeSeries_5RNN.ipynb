## 순환신경망(RNN Recurrent neural network)
- 유닛간의 연결이 순환적 구조를 갖는 신경망

- 일반적인 신경망
    - 각각의 입출력이 서로 독립적
    - 입력층에서 출력층 한 방향으로만 흐르는 feed forward 신경망

- 기존 신경망과의 차이점
    - 내부의 메모리를 이용해 시퀀스 형태의 입력을 처리할 수 있음
    - 순서가 있는 정보를 입력 데이터로 사용한다는 점이 특징

- 응용 분야
    - 필기체 인식, 자연어 처리(NL, Natural Language), 음성 인식, 작곡, 동영상 분석, 주가예측 등과 같은 순차적인(sequential) 시계열(time series) 데이터에 적합한 모델

- 순차데이터 - 샘플에 순서가 있는 데이터 예) 문장

- 타임 스텝 - 순차 데이터를 처리하는 각 단계
    good morning => 단어라면 타임 스텝 2, 글자라면 타임 스텝 12
