## 시계열 분석(Time Series Analysis)
- 주어진 시계열 데이터를 사용하여 기계학습 모형을 만들어서 미래에 일어날 일들을 예측하는 것

- 시계열 데이터 : 시간의 흐름에 따라 관찰된 데이터
- 시계열 분석의 목적 : 과거를 분석하여 미래를 예측하기 위한 것
- 시계열 분석의 가정 : 오늘의 데이터가 내일의 데이터에 영향을 줌

- 회귀 분석과의 차이점
    - 변수가 자기상관성(Autocorrelation)을 가지는 것을 전제로 함
    - 회귀분석과 달리 시계열은 순서가 중요함
- 자기상관성: 시간의 변화에 따른 변수간의 상관관계( t 시점의 데이터가 (t-1) 시점, (t-2) 시점, (t-n) 시점의 데이터와 얼마나 관계가 있는가? )

- 정상성(stationarity) 
    - 시간에 따라 확률적인 성분(평균, 분산)이 변하지 않는다는 특징, 평균과 분산이 안정되어 있어서 분석하기 쉬움
    - 정상성을 만족하지 않는 경우 분석하기 어려우므로 정상성을 갖도록 전처리를 해야 함(일반적으로 평균이 일정하지 않으면 차분을 취하고 분산이 일정하지 않으면 로그 변환 등을 하게 된다.)
    - 안정적인 시계열 : 시간의 추이와 관계없이 평균,분산이 일정한 경우

- 불안정적인 시계열 : 평균이 일정하지 않은 시계열은 차분(difference)을 통해 안정적으로 변환한 후 분석
    - 차분(difference) : 현시점 자료에서 전시점 자료를 뺀 값
        - 1차 차분 : 현시점 자료에서 직전 시점의 자료를 뺀 값
        - 2차 차분 : 현시점 자료에서 전전 시점의 자료를 뺀 값

- AR모형(자기상관모형, Autocorrelation) : 이전의 값이 이후의 값에 영향을 미치는 모형
    - AR(1) : 직전 데이터가 다음 데이터에 영향을 줄 경우
    - 자기상관함수 ( autocorrelation function : ACF )
        - 시간의 진행에 따른 자기 상관의 정도를 나타내는 함수, 자기와 자기 이외의 시차 간의 상관계수를 구하는 함수, yt와 yt-1의 상관관계를 구해서 yt-2 이상의 시차 상관계수를 유추하는 것
        - 이 수치가 높으면 전 시차의 영향을 계속 받기 때문에 데이터가 독립적이지 않은 것을 의미함 => 불안정한 데이터 => 안정적인 시계열로 변환하는 작업이 필요함
    - 부분자기상관함수 ( partial autocorrelation function : PACF )
        - 세 개의 변수 X, Y, Z 가 있을 때 Z 의 효과를 제거한 X와 Y 간의 부분상관관계, yt와 yt-1만의 상관계수를 구하는 것, yt-2 이상의 시차는 고려하지 않음

- MA모형(이동평균, Moving Average) : 시간이 지날수록 평균값이 지속적으로 증가하거나 감소하는 경향
    - MA(1) 직전 데이터가 다음 데이터에 영향을 주는 경우

- ARIMA 모형: AR모형과 MA모형을 결합한 모형
    - arima(p,d,q)
        - p : ar(p) 모형의 p  
        - d : 차분차수
        - q : ma(q) 모형의 q
