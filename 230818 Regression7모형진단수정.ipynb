```python
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
boston
```

```python
dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])
df = pd.concat([dfX, dfy], axis=1)
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x = "LSTAT", y="MEDV", data=df)   
# LSTAT 인구 하위계층
plt.show()
```

```python
import statsmodels.api as sm
model1 = sm.OLS.from_formula("MEDV ~ LSTAT", data = df)
result1 = model1.fit()
result1.summary() 
```

```python
y_hat1 = result1.predict(df)
df1 = pd.concat([y_hat1, df.LSTAT], axis = 1).sort_values("LSTAT")
df1.columns = ['prediction', 'LSTAT']
df1.plot(x = 'LSTAT', style = 'r-', lw=3)
plt.plot(df.LSTAT, df.MEDV, 'bo', alpha = 0.5)
plt.show()   # 예측력이 약함
```

```python
# LSTAT을 제곱한 비선형 독립변수를 추가한 모형
model2 = sm.OLS.from_formula('MEDV ~ LSTAT + I(LSTAT**2)', data = df)
result2 = model2.fit()
result2.summary()   # R^2 향상됨 54% >> 64%
```

```python
y_hat2 = result2.predict(df)
df2 = pd.concat([y_hat2, df.LSTAT], axis = 1).sort_values("LSTAT")
df2.columns = ['prediction', 'LSTAT']
df2.plot(x = 'LSTAT', style = 'r-', lw=3)
plt.plot(df.LSTAT, df.MEDV, 'bo', alpha = 0.5)
plt.show()   # 선형보다 조금 복잡한 모형임
```

```python
import numpy as np
df['CAT_RM'] = np.round(df.RM)
sns.barplot(x = 'CAT_RM', y='MEDV', data=df)    # RM을 카테고리형으로 변형해봄
```

```python
#실수형을 정수형으로 바꾸고 더미변수를 사용한 모형
model4 = sm.OLS.from_formula("MEDV ~ C(np.round(RM))+0", data=df)
result4 = model4.fit()
result4.summary()   # 성능이 개선됨 48.4% => 53.7%
```

```python
# 일자별 전력수요량
# 독립변수가 시간인 경우 특정 시점에서 경과된 시간값으로 변형해야 함
data = pd.read_csv('Data/energy.csv')
data.head()
```

```python
import datetime as dt
data['Date'] = pd.to_datetime(data['Date'])
# toordinal 서기 1년 1월 1일을 기준으로 경과한 날짜
data['Ordinal'] = data.Date.map(dt.datetime.toordinal)
data['Timestamp'] = data.Date.map(dt.datetime.timestamp)
data.head()
```

```python
#일단위 날짜를 기준으로 분석, 날짜는 단위가 크므로 스케일링이 필요함
model5 = sm.OLS.from_formula("Demand ~ scale(Ordinal)",data=data)
result5 = model5.fit()
result5.summary()
```

```python
data['Year'] = data.Date.dt.year
data['Month'] = data.Date.dt.month
data['DayOfYear'] = data.Date.dt.dayofyear  # 연중 몇일 차
data['DaysInMonth'] = data.Date.dt.daysinmonth  # 일수
data['DayOfWeek'] = data.Date.dt.dayofweek  
# data['WeekOfYear'] = data.Date.dt.weekofyear  # 연중 몇주 차
data['Weekday'] = data.Date.dt.weekday  # 요일코드
data['IsMonthStart'] = data.Date.dt.is_month_start
data['IsMonthEnd'] = data.Date.dt.is_month_end
data.tail()
```

```python
data.Date.dt.weekofyear 
```

```python
formula = '''Demand ~ scale(Ordinal) + C(Month)+0 + DayOfYear +
C(DaysInMonth)+0 + C(DayOfWeek)+0 + C(Weekday)+0 +
C(IsMonthStart)+0 + C(IsMonthEnd)+0'''

model6 = sm.OLS.from_formula(formula, data=data)
result6 = model6.fit()
result6.summary()   # 3.1%에서 53.7%로 향상됨
```

```python
#독립변수 뿐 아니라 상황에 따라서는 종속변수도 변형할 수 있음
#실제주택가격과 예측값의 비교
plt.scatter(boston.target, y_hat1)
plt.xlabel('Real')
plt.ylabel('Predict')
plt.show()
#선형적으로 설명하기 어려운 모형
```

```python
#종속변수를 제곱근 처리
model11 = sm.OLS.from_formula("np.sqrt(MEDV) ~ LSTAT", data=df)
result11 = model11.fit()
result11.summary()
```

```python
plt.subplot(121)
plt.scatter(boston.target, y_hat1)
plt.title("MEDV ~ LSTAT")
plt.subplot(122)
plt.scatter(boston.target, (result11.predict(df))**2)
plt.title("np.sqrt(MEDV) ~ LSTAT")
plt.tight_layout()
plt.show()
```

```python
#로그를 취한 모형
model12 = sm.OLS.from_formula("np.log(MEDV) ~ LSTAT", data=df)
result12 = model12.fit()
print(result12.summary())
```

```python
plt.subplot(121)
plt.scatter(boston.target, y_hat1)
plt.title("MEDV ~ LSTAT")
plt.subplot(122)
plt.scatter(boston.target, np.exp(result12.predict(df)))
plt.title("np.log(MEDV) ~ LSTAT")
plt.tight_layout()
plt.show()
```

```python
# 독립변수와 종속변수 모두 로그를 취한 모형(가장 성능이 좋은 모형)
model13=sm.OLS.from_formula("np.log(MEDV) ~ np.log(LSTAT)",data=df)
result13 = model13.fit()
print(result13.summary())
```

```python
plt.subplot(121)
plt.scatter(boston.target, y_hat1)
plt.title("MEDV ~ LSTAT")
plt.subplot(122)
plt.scatter(boston.target, np.exp(result13.predict(df)))
plt.title("np.log(MEDV) ~ np.log(LSTAT)")
plt.tight_layout()
plt.show()
```
