```python
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
boston
```

```python
dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])
df = pd.concat([dfX, dfy], axis=1)
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x = "LSTAT", y="MEDV", data=df)   
# LSTAT ì¸êµ¬ í•˜ìœ„ê³„ì¸µ
plt.show()
```

```python
import statsmodels.api as sm
model1 = sm.OLS.from_formula("MEDV ~ LSTAT", data = df)
result1 = model1.fit()
result1.summary() 
```

```python
y_hat1 = result1.predict(df)
df1 = pd.concat([y_hat1, df.LSTAT], axis = 1).sort_values("LSTAT")
df1.columns = ['prediction', 'LSTAT']
df1.plot(x = 'LSTAT', style = 'r-', lw=3)
plt.plot(df.LSTAT, df.MEDV, 'bo', alpha = 0.5)
plt.show()   # ì˜ˆì¸¡ë ¥ì´ ì•½í•¨
```

```python
# LSTATì„ ì œê³±í•œ ë¹„ì„ í˜• ë…ë¦½ë³€ìˆ˜ë¥¼ ì¶”ê°€í•œ ëª¨í˜•
model2 = sm.OLS.from_formula('MEDV ~ LSTAT + I(LSTAT**2)', data = df)
result2 = model2.fit()
result2.summary()   # R^2 í–¥ìƒë¨ 54% >> 64%
```

```python
y_hat2 = result2.predict(df)
df2 = pd.concat([y_hat2, df.LSTAT], axis = 1).sort_values("LSTAT")
df2.columns = ['prediction', 'LSTAT']
df2.plot(x = 'LSTAT', style = 'r-', lw=3)
plt.plot(df.LSTAT, df.MEDV, 'bo', alpha = 0.5)
plt.show()   # ì„ í˜•ë³´ë‹¤ ì¡°ê¸ˆ ë³µìž¡í•œ ëª¨í˜•ìž„
```

```python
# ì¼ìžë³„ ì „ë ¥ìˆ˜ìš”ëŸ‰
# ë…ë¦½ë³€ìˆ˜ê°€ ì‹œê°„ì¸ ê²½ìš° íŠ¹ì • ì‹œì ì—ì„œ ê²½ê³¼ëœ ì‹œê°„ê°’ìœ¼ë¡œ ë³€í˜•í•´ì•¼ í•¨
data = pd.read_csv('Data/energy.csv')
data.head()
```

```python
import datetime as dt
data['Date'] = pd.to_datetime(data['Date'])
# toordinal ì„œê¸° 1ë…„ 1ì›” 1ì¼ì„ ê¸°ì¤€ìœ¼ë¡œ ê²½ê³¼í•œ ë‚ ì§œ
data['Ordinal'] = data.Date.map(dt.datetime.toordinal)
data['Timestamp'] = data.Date.map(dt.datetime.timestamp)
data.head()
```

```python
#ì¼ë‹¨ìœ„ ë‚ ì§œë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ì„, ë‚ ì§œëŠ” ë‹¨ìœ„ê°€ í¬ë¯€ë¡œ ìŠ¤ì¼€ì¼ë§ì´ í•„ìš”í•¨
model5 = sm.OLS.from_formula("Demand ~ scale(Ordinal)",data=data)
result5 = model5.fit()
result5.summary()
```

```python
data['Year'] = data.Date.dt.year
data['Month'] = data.Date.dt.month
data['DayOfYear'] = data.Date.dt.dayofyear  # ì—°ì¤‘ ëª‡ì¼ ì°¨
data['DaysInMonth'] = data.Date.dt.daysinmonth  # ì¼ìˆ˜
data['DayOfWeek'] = data.Date.dt.dayofweek  
# data['WeekOfYear'] = data.Date.dt.weekofyear  # ì—°ì¤‘ ëª‡ì£¼ ì°¨
data['Weekday'] = data.Date.dt.weekday  # ìš”ì¼ì½”ë“œ
data['IsMonthStart'] = data.Date.dt.is_month_start
data['IsMonthEnd'] = data.Date.dt.is_month_end
data.tail()
```

```python
data.Date.dt.weekofyear 
```

```python
formula = '''Demand ~ scale(Ordinal) + C(Month)+0 + DayOfYear +
C(DaysInMonth)+0 + C(DayOfWeek)+0 + C(Weekday)+0 +
C(IsMonthStart)+0 + C(IsMonthEnd)+0'''

model6 = sm.OLS.from_formula(formula, data=data)
result6 = model6.fit()
result6.summary()   # 3.1%ì—ì„œ 53.7%ë¡œ í–¥ìƒë¨
```

```python
#ë…ë¦½ë³€ìˆ˜ ë¿ ì•„ë‹ˆë¼ ìƒí™©ì— ë”°ë¼ì„œëŠ” ì¢…ì†ë³€ìˆ˜ë„ ë³€í˜•í•  ìˆ˜ ìžˆìŒ
#ì‹¤ì œì£¼íƒê°€ê²©ê³¼ ì˜ˆì¸¡ê°’ì˜ ë¹„êµ
plt.scatter(boston.target, y_hat1)
plt.xlabel('Real')
plt.ylabel('Predict')
plt.show()
#ì„ í˜•ì ìœ¼ë¡œ ì„¤ëª…í•˜ê¸° ì–´ë ¤ìš´ ëª¨í˜•
```

```python
#ì¢…ì†ë³€ìˆ˜ë¥¼ ì œê³±ê·¼ ì²˜ë¦¬
model11 = sm.OLS.from_formula("np.sqrt(MEDV) ~ LSTAT", data=df)
result11 = model11.fit()
result11.summary()
```

```python
plt.subplot(121)
plt.scatter(boston.target, y_hat1)
plt.title("MEDV ~ LSTAT")
plt.subplot(122)
plt.scatter(boston.target, (result11.predict(df))**2)
plt.title("np.sqrt(MEDV) ~ LSTAT")
plt.tight_layout()
plt.show()
```

```python
#ë¡œê·¸ë¥¼ ì·¨í•œ ëª¨í˜•
model12 = sm.OLS.from_formula("np.log(MEDV) ~ LSTAT", data=df)
result12 = model12.fit()
print(result12.summary())
```

```python
plt.subplot(121)
plt.scatter(boston.target, y_hat1)
plt.title("MEDV ~ LSTAT")
plt.subplot(122)
plt.scatter(boston.target, np.exp(result12.predict(df)))
plt.title("np.log(MEDV) ~ LSTAT")
plt.tight_layout()
plt.show()
```
