BLUR

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('c:/workspace/Data/winter.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

```python
dst = cv2.blur(img, (9,9))   # kernel 사이즈(값이 클수록 흐려짐)
plt.imshow(dst)
```

```python
dst_rgb = cv2.blur(img_rgb, (9,9))   # kernel 사이즈(값이 클수록 흐려짐)
plt.imshow(dst_rgb)
```

```python
plt.figure(figsize=(10,7))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('original')
plt.axis('off')

plt.subplot(1,2,2)
img_blr = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(img_blr)
plt.title('dst')
plt.axis('off')

plt.show()
```
