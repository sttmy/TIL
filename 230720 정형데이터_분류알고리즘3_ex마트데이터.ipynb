```python
train_cols = ['seoul', 'incheon', 'gyungi', 'male', 'female', 'age10', 'age2030',
              'age4050', 'age60', 'food', 'drink', 'product', 'max_ta', 'max_ws',
              'min_ta', 'avg_ta', 'avg_rhm', 'avg_ws', 'sum_rn']
```

```python
# under sampling
from imblearn.under_sampling import RandomUnderSampler
x_sample, y_sample = RandomUnderSampler(random_state=0).fit_resample(x, y)
```

```python
x_sample = pd.DataFrame(data= x_sample, columns = train_cols)
y_sample = pd.DataFrame(data= y_sample, columns = ['class'])
df_sample = pd.concat([x_sample, y_sample], axis = 1)
y_sample.value_counts()
```

```python
X = x_sample[train_cols]
Y = y_sample['class']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# pipeline: 작업 순서
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(random_state=0)),
]) 
params = {
    'clf__C': [0.0001, 0.001, 0.01, 0.1, 1.0, 10, 100],  
}
grid = GridSearchCV(pipe, params, scoring = 'accuracy', cv = 3)
grid.fit(X, Y)
print(grid.best_params_)
print(grid.best_score_)
```

```python
# 다양한 알고리즘별 정확도를 저장할 리스트

scores=[]
scores.append(grid.best_score_)
print(scores)
```

```python
##### 의사결정나무
from sklearn.tree import DecisionTreeClassifier
pipe = Pipeline([
    ('scaler', StandardScaler()),  
    ('clf', DecisionTreeClassifier(random_state=0)),
])
params = {
     'clf__max_depth': list(range(3,6))  
}
grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, Y)
print(grid.best_params_)
print(grid.best_score_)
scores.append(grid.best_score_)
print(scores)
```

```python
tree = grid.best_estimator_['clf']
```

```python
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
from IPython.display import Image

dot_data = export_graphviz(tree, out_file=None, feature_names=train_cols, 
                           class_names=["0","1"], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())

# 트리를 만드는 결정에 각 특성이 얼마나 중요한지를 평가하는 특성 중요도 적용
# 0~1 사이의 값(0 전혀 사용되지 않음 ~ 1 완벽하게 타겟 클래스 예측)
```

```python
import numpy as np
def plot_tree(model):
    n_features = X.shape[1]
    plt.rcParams["figure.figsize"]=(16,9)
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), train_cols)
    plt.xlabel("importances")
    plt.ylabel("features")
    plt.ylim(-1, n_features)
plot_tree(tree)
```

```python
##### 랜덤포레스트

from sklearn.ensemble import RandomForestClassifier
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', RandomForestClassifier(random_state=0)),
])

params = {
        'clf__max_depth': list(range(3,6)),
        'clf__n_estimators': list(range(10,110,10)),
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, Y)
print(grid.best_params_)
print(grid.best_score_)
scores.append(grid.best_score_)
print(scores)
```

```python
rf = grid.best_estimator_['clf']
plot_tree(rf)
```

```python
##### KNN
from sklearn.neighbors import KNeighborsClassifier
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', KNeighborsClassifier()),
])
params = {
       'clf__n_neighbors': list(range(1,11)),
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, Y)
print(grid.best_params_)
print(grid.best_score_)
scores.append(grid.best_score_)
print(scores)
```

```python
##### sklearn 신경망: ANN

from sklearn.neural_network import MLPClassifier
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', MLPClassifier(random_state=0, max_iter=5000)),
])
params = {
        'clf__hidden_layer_sizes': [(50,), (100,)],
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, Y)
print(grid.best_params_)
print(grid.best_score_)
scores.append(grid.best_score_)
print(scores)
```

```python
# SVM

from sklearn.svm import SVC

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', SVC(random_state=0)),
])
params = {
        'clf__C': [0.01, 0.1],
        'clf__gamma': [0.001, 0.01],
}

grid = GridSearchCV(pipe, params, scoring='accuracy', cv=3)
grid.fit(X, Y)
print(grid.best_params_)
print(grid.best_score_)
scores.append(grid.best_score_)
print(scores)
```

```python
from sklearn.neural_network import MLPClassifier

pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('clf', MLPClassifier(random_state=0, max_iter=5000))
])
params={
    'clf__hidden_layer_sizes': [(50,), (100,)],
    'clf__activation': ["tanh", "relu"],
    'clf__alpha': [0.001, 0.01, 0.1]
    }

grid=GridSearchCV(pipe,params,scoring='accuracy',cv=3)
grid.fit(X, Y)
grid.best_params_
scores.append(grid.best_score_)
print(scores)
```

```python
##### 케라스 신경망: DNN

from scikeras.wrappers import KerasClassifier
from keras import models
from keras import layers
def create_network(optimizer="rmsprop"):
    network = models.Sequential()
    network.add(layers.Dense(units=16,
                         activation="relu",
                         input_shape=(X.shape[1],)))
    network.add(layers.Dense(units=16, activation="relu"))
    network.add(layers.Dense(units=1, activation="sigmoid"))    
    network.compile(loss='binary_crossentropy',
                    optimizer=optimizer, # 옵티마이저
                    metrics=["accuracy"]) # 성능 지표    

    return network
```

```python
neural_network = KerasClassifier(model=create_network, verbose=0)

# 하이퍼파라미터 탐색 영역을 정의
pipe=Pipeline([
    ('scaler',StandardScaler()),
    ('clf',neural_network)
])
params={
    'clf__epochs': [50, 100],
    'clf__batch_size': [32,64,128],
    'clf__optimizer': ["rmsprop", "adam"],
    }

grid=GridSearchCV(pipe,params,scoring='accuracy',cv=3)
grid_result = grid.fit(X, Y)
grid_result.best_params_
scores.append(grid.best_score_)
print(scores)
```

```python
##### 알고리즘별 예측도 그래프
methods=["Logistic","Tree","RF","KNN","ANN","SVM","DNN"]
scores_arr=np.array(scores) #검증용 데이터셋의 정확도
max_rate=np.max(scores_arr) #가장 좋은 정확도
idx=np.argmax(scores_arr) #최대값의 인덱스
print(methods)
print(scores_arr)
print("최적의 모형:",methods[idx])
print("최고 정확도:",max_rate)       

plt.rcParams["font.size"]=15
plt.bar(methods, scores_arr)
plt.xlabel("Classifier")
plt.ylabel("Accuracy")
```
