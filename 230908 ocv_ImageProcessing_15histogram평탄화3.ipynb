image pixel 히스토그램 평탄화 (평활화)

- 치우쳐져 있는 픽셀의 히스토그램, 평탄하게 맞춤

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('c:/workspace/Data/santorini.png')
print(img.shape, img[0])
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape, gray[0])
plt.imshow(gray,cmap='gray')
plt.axis("off")
plt.show()
```

```python
h=cv2.calcHist([gray],[0],None,[256],[0,256])    
plt.plot(h,color='r',linewidth=1)
plt.show()
```

```python
# 히스토그램 평활화

equal=cv2.equalizeHist(gray)            
plt.imshow(equal,cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
h=cv2.calcHist([equal],[0],None,[256],[0,256])    
plt.plot(h,color='r',linewidth=1)
plt.show()
```
