```python
lis = list.find_all('li')
dt = []
for li in lis:
    rank_spans = li.select('div > span.rankBtn_ctrl')
    rank = ''
    for span in rank_spans:
        rank += span['class'][1][-1]
#     title = li.select_one('div > strong').string.strip()
    url_add = li.select_one('a')['href']
    base_url = 'https://book.interpark.com'
    url = base_url+url_add
    response = requests.get(url)
    sublist = BeautifulSoup(response.text, 'html.parser')

    sli = sublist.select('#inc_optionWrap > div.optionRight_wrap > div.bookInfoBox > ul > li')
    authors = sli[0].get_text().split(": ")[1]
    company = sli[1].get_text().split(": ")[1]
    issuedate = sli[2].get_text().split(": ")[1]
    pages = sli[3].get_text().split(": ")[1]
    price = sublist.select('p.fixedPrice')[0].text
    
    dt.append([rank, authors, company, issuedate, pages, price])
```
