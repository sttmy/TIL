### 3. 사용시간 단위 통일

```python
def convert_time_minute(time_list): 
    for time in time_list:
        if '~' in time:
            t = time.split('~')[0]
        if '시간' & '분' in time:
            t = int(time.split('시간')[0])*60 + int(time.split('시간')[1].replace('분',''))
        if '시간' in time:
            t = int(time.split('시간')[0])*60
        if '분' in time:
            t = int(time)
        else:
            t = time
    return t
```

```python
# tch

def convert_time_minute(time):
    try:
        if '시간' in time:   # 시간 또는 시간+분 있는 경우
            hour = time.split('시간')[0]  # 1시간 20분 >> 1
            if '분' in time:
                minute = time.split('시간')[-1].split('분')[0]  # 1시간20분 >> 20
            else:
                minute = 0
        else:                # 분만 있는 경우
            hour = 0
            minute = time.split('분')[0]
        return int(hour) * 60 + int(minute)
    except:
        return None
```

```python
# 동작 테스트
times = ['45분', '1시간', '9~30분', '1시간', '3시간30분','2시간','8분']
for time in times:
    time_value = convert_time_minute(time)
    print(time , '=', time_value, '분')
```

```python
new_use_time_list = []
for time in use_time_list:
    value = convert_time_minute(time)
    new_use_time_list.append(value)
len(new_use_time_list), new_use_time_list[:10]
```

### 4. 무선청소기 흡입력 단위 통일

```python
# W는 지우고 pa는 100으로 나누어서 단위를 통일
# 소문자 있을 수 있고, AW, ~ 등

def get_suction(value):
    try:
        value = value.upper()
        if 'AW' in value or 'W' in value:
            result = value.replace('A','').replace('W','')
            result = int(result.replace(',',''))
        elif "PA" in value:
            result = value.replace("PA",'')
            result = int(int(result.replace(',','')) / 100)
        else:
            result = None
        return result
    except:
        return None
```

```python
suc = ['15,000Pa', '220W', '13,000Pa', '220W', '210W']

for su in suc:
    suc_va = get_suction(su)
    print(suc_va)
```

```python
new_suction_list = []
for power in suction_list:
    value = get_suction(power)
    new_suction_list.append(value)
len(new_suction_list), new_suction_list[10:20]
```
