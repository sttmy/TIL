### 1. 셀레늄 + BS

```python

import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome()
url = 'https://www.melon.com/chart/' 
# url = 'https://www.melon.com/chart/htm' 
driver.get(url)
html = driver.page_source   # 페이지소스를 가져옴
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.select('tr')[1:]
```

```python
songs = soup.select('tr')[1:]
```

```python
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, singer)
```

### 2. Selenium만 사용

```python
# Selenium만 사용해봄

url = 'https://www.melon.com/chart/' 
driver.get(url)
songs = driver.find_elements("css selector",'tr')[1:]
songs
```

```python
for song in songs:
    title = song.find_elements("css selector", 'div.ellipsis.rank01 > span > a')[0].text
    singer = song.find_elements("css selector", 'div.ellipsis.rank02 > a')[0].text
    print(title, singer, sep = " | ")
```
