## 머신러닝 알고리즘 개요

### 인공지능(AI)

- 학습, 문제 해결, 패턴 인식 등과 같이 주로 인간 지능과 연결된 인지 문제를 해결하는 데 주력하는 컴퓨터 공학 분야

- 1956년 미국 다트머스 대학에서 열린 회의에서 존 매카시는 처음으로 인공지능(AI)이란 용어를 언급하며 “기계를 인간 행동의 지식에서와 같이 행동하게 만드는 것”이라고 정의함

- 또한, 영국의 수학자인 앨런 튜링은 컴퓨터가 사람처럼 생각할 수 있다는 견해를 제시하며, 대화를 나누어 상대방이 컴퓨터인지 사람인지를 구별할 수 없다면 그 컴퓨터는 사고할 수 있는 것으로 간주해야 한다고 주장함. 이 이론은 지금까지 인공지능 분야의 기반이 되었으며, 튜링 테스트(Turing Test)라는 이름으로 인공지능을 판별하는 기준으로 활용되고 있음

- 인공지능을 구현하는 대표적인 방법 중 하나가 바로 머신러닝이고, 딥러닝은 머신러닝의 여러 방법 중 하나의 방법론으로 인공신경망(Artificial Neural Network)의 한 종류임

### 머신러닝(Machine Learning)

- 머신러닝은 컴퓨터를 인간처럼 학습시킴으로써 컴퓨터가 새로운 규칙을 생성할 수 있지 않을까 하는 시도에서 시작되어 컴퓨터가 스스로 학습할 수 있도록 도와주는 알고리즘이나 기술을 개발하는 분야
- 알고리즘을 이용하여 데이터를 분석하고, 분석 결과를 스스로 학습한 후, 이를 기반으로 어떠한 판단이나 예측을 하는 것을 의미함. 따라서 머신러닝에서는 양질의 데이터가 매우 중요한 역할을 하며, 양질의 데이터를 많이 보유할수록 보다 높은 성능을 이끌어낼 수 있게 됨. 
- 이러한 양질의 데이터를 얻기 위해 데이터 분석 과정 중에 데이터 정제 과정을 거침

- 머신러닝(Machine Learning)은 2000년대에 들어서 시작된 인공지능을 구현하는 방법의 하나로 빅데이터를 스스로 분석하고,  그 내용을 바탕으로 결론을 도출하는 기술
    - 빅데이터란, 기존 데이터베이스 관리도구의 능력을 넘어서는 대량(수십 테라 ~ 페타바이트)의 정형 또는 비정형의 데이터임. 기술 향상으로 컴퓨터 시스템이 발전하면서 시스템을 개발, 운영하는 과정에서 수많은 데이터들이 쌓였고, 인공지능 개발자들은 이 속에서 전문가와 같은 고급 지식이 아니더라도, 데이터로부터 사람 수준의 간단한 지능을 구현할 방법을 고민하게 되었고, 데이터에 사람이 생각하는 정답(label)을 매겨서 계산하고, 사람의 생각과 차이가 나는 오류를 줄여가는 방법으로 수정하고, 이러한 과정을 반복함으로써 사람의 생각과 유사하게 만들 수 있을 것이라는 방법을 생각하게 됨. 이러한 과정을 학습(training)이라고 부르고, 학습 결과로 얻어진 지능을 실제 응용에 적용하는 이 기술을 바로 “머신러닝(Machine Learning)”이라고 부름
    
### 머신러닝 학습방법

#### 1) 지도 학습(Supervised Learning)
- 입력 값과 함께 결과 값(정답 레이블)을 같이 주고 학습을 시키는 방법
- 예) 정답을 기반으로 오류를 줄여서 학습하는 방법 즉, 반복 학습을 통해 오류를 줄여가면서 점차 정답에 가까워지는 방법
- 대표적인 알고리즘: 회귀(Regression)알고리즘,  분류(Classification)알고리즘, 부스팅알고리즘 

#### 2) 비지도 학습(Unsupervised Learnging)
- 지도 학습과 달리 정답을 알려주지 않고 예측하는 방법
- 정답을 모르더라도 유사한 것들과 다른 것들을 구분해서 군집을 만들 수 있는 학습 방법
- 예) 아이가 자라면서 처음 보는 장난감도 엄마의 화장품이나 맛있는 음식과 다르다는 것을 구분할 수 있게 되고, 많은 사람이 말하더라도 엄마의 목소리를 구분해 낼 수 있는 것과 같은 방법
- 대표적인 알고리즘: 군집(클러스터링,Clustering)알고리즘

#### 3) 강화 학습(Reinforcement Learning)
- 보상 및 벌칙과 함께 여러 번의 시행착오를 거쳐 스스로 학습하는 방법
- 예) 아이가 스스로 일어서는 방법, 스스로 걷는 방법 등은 엄마가 있더라도 정답을 알려주기 어려움. 스스로 넘어져 가면서 여러 번의 시행착오 끝에 자기에게 맞는 방법을 습득하게 되는 것
- 지도학습, 비지도 학습과 달리 보상과 벌칙을 반영해 반복 시행을 통해서 데이터를 스스로 만들어가면서 학습
- 많은 학습 시간이 필요해 컴퓨터 시뮬레이션 기반으로 빠르게 반복함으로써 이러한 학습시간의 한계점을 극복해 가고 있습니다.
- 대표적인 알고리즘: 딥마인드의 알파고

### 모델 평가요소 (평가지표)

- 모델 평가 함수: evaluate() 
    - 모델의 최종적인 정답률과 loss값을 알 수 있음
    - loss 예측값과 실제값이 차이나는 정도를 나타내는 지표. 작을수록 좋음

- 정확도(Accuracy): 실제 정답과 얼마나 유사한지

- 정밀도(Precision): 정답이라 예측한 것들 중에서 실제 정답인 것들의 비율. 정밀도가 높다고 정확하다는 의미는 아님

- 재현율(Recall): 실제 정답인 것들 중에서 예측모델이 정답이라 예측한 것의 비율

- F1 Score: 정밀도와 재현율의 조화평균. 

- 민감도와 특이도 

- F1스코어는, 개체명인식에 주로 사용됨. 범주가 불균형한 상태에서 사용 (?)
