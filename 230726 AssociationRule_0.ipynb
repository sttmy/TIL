## 연관규칙 분석

### A. 연관규칙분석(Association Rule Mining)
- 1) 소비자 또는 일련의 사용자 그룹의 행동 간에 존재하는 동시발생 관계를 분석하는 알고리즘
- 2) 대규모 데이터베이스에 존재하는 아이템 간의 관계를 규명하기 위해 사용하는 방법
- 3) 장바구니 분석(Market Basket Analysis) 또는 친밀도 분석(Affinity Analysis)이라고도 함
- 4) 맥주와 기저귀
    - ① 월마트에서 맥주와 기저귀를 비슷한 곳에 진열하니 매출이 향상됨
    - ② 남편이 마트에 들러 물건을 살 때 부인의 부탁으로 아이 기저귀를 사면서 동시에 본인이 마실 맥주를 산다는 구매 패턴을 발견하게 됨
    - ③ 데이터 분석을 통해 전혀 관련성이 없는 맥주와 기저귀의 연결고리를 알게 되어 마케팅 전략에 반영한 사례

### B. 연관 규칙 학습 알고리즘
- 1) 규칙 발견
    - ① 특정 제품을 함께 구매할 가능성: 햄버거 패티를 사는 사람은 햄버거 롤, 케첩, 머스타드, 토마토 등을 함께 사는 경향이 있다.
    - ② 상품 친밀도가 모호한 경우 : 치약과 참치캔: 일부 지역에서는 참치 요리를 한 후 즉각적으로 칫솔질하는 지역이 있을 수 있지만 일반적으로는 친밀도가 약함
    - ③ 영수증 한 장을 하나의 거래로 보고 한줄에 기록: 해당 아이템을 구매한 경우 1, 구매하지 않은 경우 0
        거래ID    양파    감자    버거    맥주     기저귀
            1       0       1       1       0       0
            2       1       1       1       1       0
            3       0       0       0       1       1
            4       1       0       1       1       0
        양파와 감자를 구매한 사람은 햄버거 패티도 살 가능성이 높다
        {양파,감자} -> {햄버거 패티}
        젊은 남성이 기저귀를 구매할 경우 맥주도 함께 살 가능성이 높다.
        {기저귀} -> {맥주} 
    - ④ 지지도(support)
        - 해당 아이템이 포함된 거래의 비율: 품목A와 품목B를 포함하는 거래수 / 전체 거래수
        - {감자,햄버거} 감자와 햄버거를 같이 산 경우는 전체 4건의 거래 중 2회이므로 50%의 거래 빈도: 2/4 = 0.5
    - ⑤ 신뢰도(confidence) : 예측의 정확성
        - 항목A의 거래 중에서 항목 B가 포함된 거래의 비율
        - {양파,햄버거} -> {맥주}의 거래 규칙은 100%의 신뢰도
    - ⑥ 향상도(Lift:Improvement)
        - A->B의 연관 규칙에서 임의로 B가 구매되는 경우에 비해 A와의 관계가 고려되어 구매되는 경우의 비율
            1 => 서로 독립적 관계, 과자와 후추
            >1 : 양의 상관관계, 빵과 버터
            <1 : 음의 상관관계, 설사약과 변비약

- 2) 예제 - 셔츠를 구매한 고객이 넥타이도 함께 구매할 연관성 분석

        고객1 : 넥타이, 셔츠, 양말

        고객2 : 양말, 벨트, 장갑, 셔츠

        고객3 : 지갑, 넥타이, 셔츠
        
        고객4 : 양말, 벨트, 장갑, 바지
    
    P(A):셔츠 구매, P(B) 넥타이 구매
    
    동시 거래수(셔츠, 넥타이) : 2회
    
        고객1 : 넥타이, 셔츠, 양말
    
        고객2 : 양말, 벨트, 장갑, 셔츠
    
        고객3 : 지갑, 넥타이, 셔츠
    
        고객4 : 양말, 벨트, 장갑, 바지
    
    셔츠 구매 거래수 : 3회
        고객1 : 넥타이, 셔츠, 양말
    
        고객2 : 양말, 벨트, 장갑, 셔츠
    
        고객3 : 지갑, 넥타이, 셔츠
        
        고객4 : 양말, 벨트, 장갑, 바지
    
    넥타이 구매 거래수 : 2회
    
        고객1 : 넥타이, 셔츠, 양말
    
        고객2 : 양말, 벨트, 장갑, 셔츠
    
        고객3 : 지갑, 넥타이, 셔츠
    
        고객4 : 양말, 벨트, 장갑, 바지
    
    전체 거래수 4회
    
    개별확률 
    
        P(A) 셔츠를 구매할 확률 3/4 = 0.75
    
        P(B) 넥타이를 구매할 확률 2/4 = 0.5
    
    지지도 Support: A → B 전체 거래 중 A와 B가 동시에 포함된 거래의 비율, 교집합/전체
    
        셔츠와 넥타이를 동시에 구매할 확률
    
        동시거래 / 전체거래 = 2/4 = 0.5
    
    신뢰도 Confidence: A → B의 신뢰도는 품목 A를 구매했을 때 품목 B를 추가로 구매할 확률
    
        셔츠를 구매한 거래 중에서 넥타이를 구매한 확률
    
        지지도 / P(A) = 0.5 / 0.75 = 0.67
    
    향상도 Lift: A를 구매했을 때 B를 구매할 가능성
    
        셔츠->넥타이의 연관 규칙에서 임의로 넥타이가 구매되는 경우에 비해 셔츠와의 관계가 고려되어 구매되는 경우의 비율
    
        신뢰도 / P(B) = 0.67 / 0.5 = 1.34
        
        향상도가 1보다 크므로 넥타이와 셔츠는 양의 관계이며 동시에 구매할 가능성이 높음

- 3) 선험적 알고리즘(Apriori 알고리즘)
    - 거래와 관련한 빈도 패턴 마이닝, 연관 규칙 학습에 사용되는 알고리즘
    - 모든 아이템셋을 찾기보다는 아주 빈번한 아이템셋의 특성만을 찾는다.
    - 예를 들어 자동차물품과 여성용 화장품을 모두 판매할 경우 {엔진오일, 립스틱}의 집합과 같이 드물게 나타나는 조합을 무시하여 아이템셋의 크기를 줄이는 방법을 사용한다.

- 4) 빈발 패턴 성장 알고리즘(FP 성장 알고리즘, Frequent Pattern)
    - Apriori 알고리즘을 개선한 알고리즘
    - 데이터셋에서 해당 아이템의 발생 횟수를 측정한 후 문자열을 저장할 수 있는 순차적인 데이터 구조를 지닌 트리를 생성한다.
    - 다수의 거래에서 가장 빈번히 등장하는 아이템을 트리의 뿌리와 가장 가까운 요소로 나타낸다.
