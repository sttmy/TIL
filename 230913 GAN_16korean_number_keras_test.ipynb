```python
import tensorflow as tf
import pandas as pd

generator = tf.keras.models.load_model('Data/result/korean_number_gen_test.h5')
generator.summary()

```

```python
from matplotlib import pyplot as plt
import PIL
plt.figure(figsize=(10,10))
img=PIL.Image.open('image_at_epoch_0100.png')
plt.imshow(img, 'gray')
plt.title('epoch 100')
plt.axis('off')
plt.show()
```

```python
noise_dim = 100
num_ex_to_generate = 64
seed = tf.random.normal([num_ex_to_generate, noise_dim])
pred = generator(seed, training=False)
fig = plt.figure(figsize = (8,8))
for i in range(pred.shape[0]):
    plt.subplot(8,8,i+1)
    plt.imshow(pred[i,:,:,0]*127.5 + 127.5, cmap='gray')
    plt.axis('off')
```
