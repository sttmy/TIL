### 3.3 반복문

#### random 모듈
- choice(seq) 비어있지 않은 sequence에서 임의 요소 선택
- choices(population, k) 모집단으로부터 k의 크기를 갖는 목록 선택
<br>

- randint(a,b) [a,b] 범위에서 난수 정수 반환
- random(...)  [0,1] 사이의 실수 반환
- sample(population, k) 모집단으로부터 k개를 임의 추출
- seed() 씨드값 설정, 동일한 난수 선택
<br>
- uniform(a,b) [a,b] 범위의 난수 실수를 균등하게 반환
- normalvariate(mu, sigma) 정규분포 따르는 난수 반환

```python
cnt = 0
while True:
    r = random.random()
    print(r)
    if r < 0.01:
        break
    else:
        cnt += 1
cnt
```

```python
string = ''' 나는 홍길동 입니다.
주소는 서울시 입니다.
나이는 35세 입니다.'''

sent = []
words = []

for sen in string.split(sep = '\n'):
    sent.append(sen)
    for word in sen.split():
        words.append(word)

sent, len(sent), words, len(words)
```
