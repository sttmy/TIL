
#### 5.2. ProductItem.js (React)
파일 위치: c:/react/frontend/src/components/ProductItem.js
```
import React from 'react';

import './main.css'

function ProductItem({product_code, product_name, price, filename}){
    let loading = false;
    const url = `http://localhost/static/images/${filename}`;
    // `` backtick
    if (loading){
        return(
            <div>loading</div>
        )
    } else {
        let img = '';
        if (filename !== '-'){
            img = `<img src=${url} width='100px' height='100px'/><br />`;
        } else{
            img = '[상품 이미지 미등록]<br />';
        }
        return(
            <div style={{ margin: '5px'}}>
                <span dangerouslySetInnerHTML={{__html:img}}></span>
                상품명: {product_name}<br />
                가격: {price}원
                <br /><br />
            </div>
        )
    }
}

export default ProductItem;

```


#### 5.3. ListProduct.js (React)
파일 위치: c:/react/frontend/src/components/ListProduct.js
```
import React, { useEffect, useState } from 'react';
import './main.css';
import ProductItem from './ProductItem';

function ListProduct(){
    const [items, setProductList] = useState([]);

    function getList(url){
        fetch('http://localhost'+url)           // 서버호출
        .then(response => {return response.json(); })
        // 입력변수 => 출력내용 
        // function 함수이름(입력변수){ return 내용 }
        // response:   서버응답
        // response.json():   json으로 변환   (모바일, PC 등 다양한 경로로 접속, 표준 처리 포맷은 json)
        .then(data => {setProductList(data); })
    }
    useEffect(() => {getList('/list');}, []);
    // useEffect: 리엑트 컴퓨넌트가 렌더링될 때마다 특정 작업을 실행
    // 입력 () => {} 출력: function() => {return...}

    return(    //1fr : 1 fraction 1행 5열
        <>
            <h2>상품목록</h2>
            등록된 상품수: {items.length}
            <br/><br/>
            <div style={{
                display: 'grid',
                gridTemplateRows: '1fr',
                gridTemplateColumns: '1fr 1fr 1fr 1fr 1fr',
            }}>
            {items.map(
                ({ product_code, product_name, price, filename}) => (
                    <ProductItem
                    product_code={product_code}
                    product_name={product_name}
                    price={price}
                    filename={filename}
                    key={product_code} />
                )
            )}
            </div>
        </>
    );
};

export default ListProduct;
```
