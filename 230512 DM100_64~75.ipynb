67. Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값은 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의하라

```python
df.Yr_Mo_Dy.dt.year-100
```

```python
df.Yr_Mo_Dy
```

```python
df['Yr_Mo_Dy'].apply(lambda x: x.year-100)
```

```python
# datetime64에서 string으로 바꾸기

import datetime as dt

t = dt.datetime(2021, 3, 24, 11, 58, 50)
print(t.strftime('%d%m%Y'))
print(t.strftime('%Y-%m-%d %H:%M:%S'))
print(t.strftime('%y-%m-%d %H:%M:%S'))
print(t.strftime('%w'))
print(t.strftime('%U'))
```

```python
def fix_year(x):
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(dt.date(year, x.month, x.day))
df['Yr_Mo_Dy'].apply(fix_year)
```

```python
df.Yr_Mo_Dy = df.Yr_Mo_Dy.apply(lambda x: dt.date(x.year-100, x.month, x.day) if x.year >= 2061 else x)
```
