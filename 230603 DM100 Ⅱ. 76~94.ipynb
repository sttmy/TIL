## Q76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df = pd.read_csv(url)
```

```python
df.head(3)
```

```python
df.info()
```

```python
pd.to_datetime(df['(년-월-일:시)'].str.split(':')[0][0])
```

```python
dt.timedelta(days = 1)
```

```python
def change_date(x):
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour== '24':
        hour = '00:00:00'
        Date = pd.to_datetime(date + " " + hour) +dt.timedelta(days = 1)
    else: 
        hour = hour + ':00:00'
        Date = pd.to_datetime(date + " " + hour) 
    return Date

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head(3)
```

## datetime

- year: 연도
- month: 월
- day: 일
- hour: 시
- minute: 분
- second: 초
- weekday: 요일 (0:월 ~ 6:일)
- strftime: 문자열
- date: 날짜 정보
- time: 시간 정보

### strftime 문자열로 변환
- %Y: 4자리 연도
- %m: 2자리 월
- %d: 2자리 일
- %H: 24시간 
- %M: 2자리 분
- %S: 2자리 초
- %A: 영어로 된 요일
- %B: 영어로 된 월

```python
x = dt.datetime.now()
x.strftime("%A %d. %B %Y")
```

## Pivot vs Pivot_table

- Pivot :   데이터 재구조화만 가능
    
- Pivot table ;     데이터 재구조화 + 결측치 대체 (fill_value) 및 집계(aggfunc) 가능
