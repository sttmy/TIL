#### 알고리즘

알고리즘이 학습한다? Tom Mitchell, 1993

① 문제 a task

② 문제에 관한 데이터 training experience (in the form of training examples)

③ 얼마만큼 성능이 있는가 a performance measure

데이터가 늘어날수록 알고리즘 성능이 좋아지면, 알고리즘이 학습한다고 함


#### Re-programming vs Re-training : 머신러닝은 re-training (이전의 접근은 re-programming)

- Feature Engineering : 경험에 의해 어떤 feature를 사용하는 게 분류를 잘 하더라는 매뉴얼로 결정
→ 최근에는 데이터가 많이 있으면 학습을 통해 분류를 잘하더라로 shift (사람의 개입을 최소화)

vs Deductive Inference : symbolic한 AI
- inference에 의해 결론에 도달


#### (데이터에 대한) 피드백에 따라 분류

① 지도학습: agent와 무관하게 사전에 피드백, label이 있는 상태, Instructive feedback

$\{(x_n, y_n) | n = (1,..,N)\}$ 

목표: regression & classification

② 비지도학습: Unsupervised  사전에 피드백, label이 없는 상태        

$\{(x_n) | n = (1,..,N)\}$

목표: representation learning(deep neural network) & clustering 
                         
③ 강화학습: 한 일에 대해서만 피드백이 주어짐 Evaluative feedback 

$\{x_n, a_n, r_n x_{n+1}|....\}$

$x_n$: 현재 상황(statement) observation

$a_n$: agent의 action. decision

$r_n$: 결과 혹은 보상
            
시간이 오래걸려서 알파고 이전까지는 간단한 문제를 푸는 시도를 함
