```python
from sklearn.datasets import load_wine
wine = load_wine()
wine
```

```python
import pandas as pd

pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
wine.target
```

```python
x = wine.data
y = wine.target
```

```python
from sklearn.model_selection import train_test_split
x_tr, x_ts, y_tr, y_ts = train_test_split(x,y,test_size=0.2)
```

```python
import torch
# 학습용데이터를 텐서로 변환
x_tr = torch.from_numpy(x_tr).float()    # 텐서
y_tr = torch.from_numpy(y_tr).long()     # 정수

# 검증용
x_ts = torch.from_numpy(x_ts).float()
y_ts = torch.from_numpy(y_ts).long()
```

```python
# 텐서를 gpu로 옮기기
x_tr = x_tr.cuda()
y_tr = y_tr.cuda()
x_ts = x_ts.cuda()
y_ts = y_ts.cuda()
```

```python
from torch.utils.data import DataLoader, TensorDataset
# 독립변수와 종속변수의 텐서를 합침
train = TensorDataset(x_tr, y_tr)
print(train[0])
```

```python
train_loader = DataLoader(train, batch_size = 16, shuffle = True)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(13,96) 
        self.fc2 = nn.Linear(96, 72)
        self.fc3 = nn.Linear(72, 64)
        self.fc4 = nn.Linear(64, 32)
        self.fc5 = nn.Linear(32, 16)
        self.fc6 = nn.Linear(16,3)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = F.relu(self.fc5(x))
        x = self.fc6(x)
        # sofrmax 함수의 결과값에 log 적용
        # dim=0 행방향, 1 열방향
        # dim=0, row에서 최대값을 구하는 방식
        return F.log_softmax(x, dim=0)
```

```python
model = Net().cuda()
import torch.optim as optim
from torch.autograd import Variable
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr = 0.001)
```

```python
# 학습시작

for epoch in range(500):
    total_loss = 0
    for train_x, train_y in train_loader:
        # 계산 그래프 구성
        train_x, train_y = Variable(train_x), Variable(train_y)
        #텐서를 gpu로 이동시킴
        train_x=train_x.cuda()
        train_y=train_y.cuda()        

        # 경사 초기화
        optimizer.zero_grad()

        # 순전파 계산
        output = model(train_x)

        # 오차계산
        loss = criterion(output, train_y)

        # 역전파 계산
        loss.backward()

        # 가중치 업데이트
        optimizer.step()

        # 누적 오차 계산
        total_loss += loss.data

    # 50회 반복마다 누적오차 출력

    if (epoch+1) % 50 == 0:
        print(epoch+1, total_loss)
```

```python
# 계산 그래프 구성

x_ts, y_ts = Variable(x_ts), Variable(y_ts)

# 출력값 계산
result = torch.max(model(x_ts).data, 1)[1]

# 모형의 정확도 측정
# gpu에 저장된 텐서를 cpu로 이동시킴

y_ts=y_ts.cpu()
result=result.cpu()
accuracy = sum(y_ts.data.numpy() == result.numpy()) / len(y_ts.data.numpy())

# 모형의 정확도 출력
accuracy
```
