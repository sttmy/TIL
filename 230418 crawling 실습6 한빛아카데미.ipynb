```python
lines = []
for i in tqdm(range(1,11)):
    base_url = 'https://www.hanbit.co.kr/academy/books/new_book_list.html?page='+'i'
    response = requests.get(base_url)
    sp = BeautifulSoup(response.text, 'html.parser')
    list = sp.select('#container > div.new_book_list_wrap > div.sub_book_list_area > li.sub_book_list')    
    
    for li in list:
        href = li.select_one('a')['href']
        site = 'https://www.hanbit.co.kr'
        sub_url = site + href

        sub_resp = requests.get(sub_url)
        sub_sp = BeautifulSoup(sub_resp.text, 'html.parser')
        sub_list = sub_sp.select('ul.info_list > li') 

        title = sub_sp.select('h3')[2].get_text()
        price = int(sub_sp.select_one('span.pbr > strong').get_text().replace(',',''))
        author = '-'
        trans = '-'
        page = 0
        for li in sub_list:
            text = li.get_text()
            if '저자 :' in text:
                author = text.split(':')[1].strip()
            if '출간 :' in text:
                date = text.split(':')[1].strip()
            if '번역 :' in text:
                trans = text.split(':')[1].strip()
            if '페이지 :' in text:
                page = text.split(' ')[2].strip()
        lines.append([title, author, trans, date, page, price])
```

```python
df = pd.DataFrame(lines, columns = ['제목','저자','역자','발행일','쪽수','가격'])
df
```

```python
df.to_csv('data1/Crawling Hanbit Booklists.csv', index = False)
```

```python
base_url = 'https://www.hanbit.co.kr'
sub_url = '/academy/books/new_book_list.html'

url = base_url + sub_url + '?page=' + str(1)  # for문에서는 url아래로만 사용할 것
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')

lis = soup.select('li.sub_book_list')
```

```python
lins = []
for li in tqdm(lis):
    
    href = li.find('a')['href']
    sub_req = requests.get(base_url + href)
    sub_soup = BeautifulSoup(sub_req.text, 'html.parser')
    title = sub_soup.select_one('.store_product_info_box > h3').get_text().strip()
    info_list = sub_soup.select('.info_list >li')
    info_dict = {}   # dataframe으로 가져옴

    for li in info_list:
        key = li.find('strong').get_text().split(" ")[0]
        value = li.find('span').get_text().split(" ")
        info_dict[key] = value

    author = info_dict['저자']
    date = info_dict['출간']
    page = int(info_dict['페이지'][0].replace(',',''))
    price = int(sub_soup.select_one('.pbr > strong').get_text().replace(',',''))
    try: 
        translator = info_dict['번역']
    except:
        translator = '-'
    lins.append([title, author, translator, date, page, price])
```

```python
# 10페이지씩 한번에 가져오기

base_url = 'https://www.hanbit.co.kr'
sub_url = '/academy/books/new_book_list.html'

lins = []
for i in tqdm(range(1,11)):
    url = base_url + sub_url + '?page=' + str(i) 
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html.parser')
    lis = soup.select('li.sub_book_list')
    
    for li in lis:
        href = li.find('a')['href']
        sub_req = requests.get(base_url + href)
        sub_soup = BeautifulSoup(sub_req.text, 'html.parser')
        title = sub_soup.select_one('.store_product_info_box > h3').get_text().strip()
        info_list = sub_soup.select('.info_list >li')
        info_dict = {}   # dataframe으로 가져옴

        for li in info_list:
            key = li.find('strong').get_text().split(" ")[0]
            value = li.find('span').get_text().strip()
            info_dict[key] = value

        author = info_dict['저자']
        date = info_dict['출간']
        page = int(info_dict['페이지'].split(" ")[0].replace(',',''))
        price = int(sub_soup.select_one('.pbr > strong').get_text().replace(',',''))
        try: 
            translator = info_dict['번역']
        except:
            translator = '-'
        lins.append([title, author, translator, date, page, price])
```

```python
data = pd.DataFrame(lins, columns = ['제목','저자','역자','발행일','쪽수','가격'])
data
```

```python
data.to_csv('data1/Hanbit tch.csv', index = False)
```
