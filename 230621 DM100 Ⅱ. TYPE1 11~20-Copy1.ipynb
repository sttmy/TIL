#### datetime 의 2가지 타입 strftime vs strptime
- strftime : STRing Format Time 문자열로 추출
    - datetime.datetime.now().strftime('%Y년 %m월 %d일 %H시 %M분 %S초') 
- strptime : STRing Parse Time datatime 타입으로 추출
    - datetime.datetime.strptime(date, '%Y-%m-%d %H:%M:%S')
